```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import normalize
```

```python
def sine(A, omega, phi, x):
    return A * np.sin(omega * x + phi)
```

```python
x1 = sine(2, 0.2, 30, np.array(range(200)))
x2 = sine(1, 0.8, 30, np.array(range(200)))
X1 = 0.7 * x1 + 0.3 * x2
X2 = 0.2 * x1 + 0.8 * x2

fig = plt.figure(figsize=plt.figaspect(0.5))
ax = fig.add_subplot(221)
ax.plot(range(200), x1)
plt.title("source 1")
ax = fig.add_subplot(222)
ax.plot(range(200), x2)
plt.title("source 2")
plt.tight_layout()

fig = plt.figure(figsize=plt.figaspect(0.5))
ax = fig.add_subplot(223)
ax.plot(range(200), X1)
plt.title("recording 1")
ax = fig.add_subplot(224)
ax.plot(range(200), X2)
plt.title("recording 2")
plt.tight_layout()
plt.show()

X = np.vstack((X1, X2)).T
X.shape
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
learning_rate = 0.0001
w = np.random.random((2, 2))
num_epoch = 2000

for i in range(num_epoch):
    for x in X:
        w = w + learning_rate * (np.outer((1 - 2 * sigmoid(np.dot(w.T, x))), x) + np.linalg.inv(w.T))

w = normalize(w, axis=1, norm='l2')
```

```python
z = np.dot(X, w)

fig = plt.figure(figsize=plt.figaspect(0.5))
ax = fig.add_subplot(121)
ax.plot(range(200), z[:, 0])
plt.title("recording 1")
ax = fig.add_subplot(122)
ax.plot(range(200), z[:, 1])
plt.title("recording 2")
plt.tight_layout()
plt.show()
```
