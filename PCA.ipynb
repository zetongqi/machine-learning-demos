```python
from pymagnitude import *
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from numpy import linalg as LA

vectors = Magnitude("/Users/zetong/HAL/HAL-9000/glove.840B.300d.magnitude")
word_list = ["cat", "dog", "deer", "fridge", "playstation", "television", "door", "window", "table", "cake", "burger", "sandwich", "beer", "liquor", "pen", "notebook", "phone"]
X = vectors.query(word_list)


X_bar = np.mean(X, axis=0)
X_tilde = X-X_bar
X_tilde.shape
S = np.matmul(np.transpose(X_tilde), X_tilde) / X_tilde.shape[0]
v, u = LA.eig(S)
u_tilde = u[:,0:3]
results = np.matmul(X, u_tilde)
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:, 0], X[:, 1], X[:, 2])
for x, y, z, text in zip(X[:, 0], X[:, 1], X[:, 2], word_list):
    ax.text(x, y, z, text, color='red')
plt.show()
```
